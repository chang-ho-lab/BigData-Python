```python
plt.figure(figsize=(16,9))
plt.plot(area_c['구분'],area_c['2018년'])
plt.title('18년도 지역별 성범죄 현황',size=30)
plt.xlabel('지역',size=20)
plt.ylabel('성범죄 횟수',size=20)
plt.grid(alpha=0.5)
plt.xticks(rotation=45)
plt.yticks()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(area_c['구분'],area_c['2015년'])
plt.title('15년도 성범죄 현황',size=30)
plt.xlabel('지역',size=20)
plt.ylabel('성범죄 횟수',size=20)
plt.xticks(rotation=45)
plt.yticks()
plt.show()
```
