```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
location1_c = pd.read_csv('성범죄자 법정동.csv',encoding='euc-kr') #encoding='cp949'
location1_c 
```

```python
location2_c = pd.read_csv('성범죄자 행정동.csv',encoding='euc-kr') #encoding='cp949'
location2_c 
```

```python
df=pd.merge(location1_c ,location2_c)
```

```python
df
```

```python
df=df.drop_duplicates() 
```

```python
df
```

```python
df=df.drop(['시도','구군','도로명','읍면동'],axis=1) 
df
```

```python
df.info()
```

```python
df=df.iloc[0:101] 
df
```

```python
df=df.astype('int')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor       
```

```python
x=df[['고유번호','행정동코드','건물번호1']] 
y=df['건물번호2']
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y) 
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.predict(x_test)
```

```python
dt.predict([[1111984,108,47]])
```

```python
dt.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
df2=['고유번호','행정동코드','건물번호1']
```

```python
x=df[df2]
y=df['건물번호2']
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y) 
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test,y_test)
```

```python
knn.predict([[113124,7071.42853,2005]])
```

```python
from sklearn.cluster import KMeans
```

```python
df.loc[:,['고유번호','건물번호2']]
```

```python
x=df.loc[:,['고유번호','건물번호2']].values
x
```

```python
kmeans=KMeans(n_clusters=30)
```

```python
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
kmeans.predict(x)
```

```python
df['클러스터번호']=kmeans.predict(x)
df
```

```python
plt.scatter(df['고유번호'], df['건물번호2'], c=df['클러스터번호'])
```
